### bs4 라이브러리
- bs4 라이브러리 안에 있는 BeautifulSoup Class를 사용
- html로 이루어진 문자형 데이터를 pasing 작업을 하여 데이터를 쉽게 추출하기 위한 Class
    - pasing
        - 데이터의 타입을 변경(Class 안에 데이터를 대입)
- html의 TAG를 기준으로 데이터(contents)를 쉽게 추출
- 웹의 구조를 어느정도 파악을 하고 사용을 하게 되면 쉽게 접근이 가능
- Parser를 활용해서 파이썬에서 접근이 쉽게 객체의 형태(변수 + 함수)로 제공


```python
from bs4 import BeautifulSoup as bs
```

```python
# html로 이루어진 문자열을 bs Class에 대입 -> pasing
soup = bs(html_data, 'html.parser')
```

```python
soup
```

- 태그의 이름을 기준으로 검색(필터)
    - soup.태그명 : 해당 태그명의 첫번째 태그를 출력
    - soup.태그명.string : 첫번째 태그의 문자열(contents)을 출력
        - ex) `<p>test<p>` --> test
    - soup.태그명['속성'] : 첫번째 태그의 해당 속성의 값을 출력
        - ex) `<a href = "https://www.google.com">구글</a>` --> 'https://www.google.com'

```python
print(soup.a)
print(soup.a.string)
print(soup.a['href'])
```

- find()
    - HTML 데이터에서 특정 태그의 첫 번째 정보를 출력
    - find(속성명 = 속성값) : 태그들 중 해당 속성명과 속성값이 같은 첫 번째 태그의 정보를 출력
    - 결과 데이터가 TAG의 형태로 출력
- find_all()
    - HTML 데이터에서 특정 태그의 모든 정보를 출력
    - limit 매개변수 : 출력해주는 데이터의 개수를 지정
    - 결과 데이터가 TAG들의 ResultSet 형태로 지정

```python
## Tag 형태로 되어있는 데이터에서 contents를 추출
a_data = soup.find('a')
print(a_data)
print(a_data.string)
print(a_data.get_text())
```

```python
## ResultSet 형태로 되어있는 데이터에서 contents 출력
a_list = soup.find_all('a')
print(a_list)
print(a_list[0].string)
print(a_list[1].get_text())
```

```python
# for문 출력
# for + range
li = []
for i in range(len(a_list)):
    li.append(a_list[i].string)
print(li)
print(len(li))

# for + data
lli = []
for data in a_list:
    lli.append(data.get_text())
print(lli)
print(len(lli))
```

```python
a_list[-6]
```

```python
# while 조건식
cnt = 0
li2 = []
while cnt<len(a_list):
    li2.append(a_list[cnt].get_text())
    cnt+=1
print(len(li2))
print(li2)

# while 무한반복
cnt = 0
li3 = []
while True:
    try:
        li3.append(a_list[cnt].get_text())
        cnt += 1
    except:
        break
print(li3)
```

```python
# map() + 일반함수
def change(x):
    # x에 입력이 되는 데이터는? -> Tag데이터
    return x.get_text()
print(list(map(change, a_list)))

# map() + lambda
print(list(map(lambda x: x.get_text(), a_list)))
```

```python
tables = head_html.find_all('strong')
```

```python
# for문
li = []
for i in tables:
    li.append(i.get_text())
print(f"for문 활용 : {li}")

# map
print(f"map함수 활용 : {list(map(lambda x: x.get_text(), tables))}")


# while
li = []
cnt = 0
while True:
    try:
        li.append(tables[cnt].get_text())
        cnt += 1
    except:
        break
print(f"while문 활용 : {li}")
        
```

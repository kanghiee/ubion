```python
# 주어진 데이터
dic = {'1위': '명량',
       '2위': '극한직업',
       '3위': '신과함께-죄와 벌',
       '4위': '국제시장',
       '5위': '어벤져스: 엔드게임',
       '6위': '겨울왕국 2',
       '7위': '베테랑',
       '8위': '아바타',
       '9위': '서울의 봄',
       '10위': '도둑들'}

cnt2 = ['누적 1,761만명',
        '누적 1,626만명',
        '누적 1,441만명',
        '누적 1,426만명',
        '누적 1,397만명',
        '누적 1,376만명',
        '누적 1,341만명',
        '누적 1,333만명',
        '누적 1,312만명',
        '누적 1,298만명']

# 영화의 순위와 영화명을 순서대로 추출
rank = list(dic.keys())
movies = list(dic.values())

# 새로운 dic를 위한 빈 dic 생성
new_dic = {}

# 각 순위에 해당하는 영화명과 누적 관객수를 하나의 dic로 병합
for i in range(len(rank)):
    # 현재 순위, 영화명, 누적 관객수 가져오기
    current_rank = rank[i]
    current_movie = movies[i]
    current_audience = cnt2[i]
    
    # 현재 순위의 영화명과 누적 관객수를 dic에 저장
    new_dic[current_rank] = {
        '영화명': current_movie,
        '누적 관객수': current_audience
    }

# 결과 출력
print(new_dic)
```

```python
# naver 로긴(우회해서)
import requests
from bs4 import BeautifulSoup as bs
import pyperclip
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
import pandas as pd
```

```python
service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
url = 'https://www.coupang.com/np/categories/195006'
driver.get(url)

productlist = driver.find_element(By.ID,'productList')
list = productlist.find_elements(By.TAG_NAME,'li')

```

```python
prd_name = []
prd_price = []
prd_deli = []
prd_star = []
prd_url = []
prd_foodname = []
next_element = driver.find_element(By.CLASS_NAME,'page-warpper')
next_page_button = next_element.find_elements(By.TAG_NAME,'a')
for idx, li in enumerate(list):
    try:
        prd_name.append(li.find_element(By.CLASS_NAME,'name').text)
        prd_price.append(li.find_element(By.CLASS_NAME,'sale').text)
        prd_deli.append(li.find_element(By.CLASS_NAME,'delivery').text)
        prd_star.append(li.find_element(By.CLASS_NAME,'rating-star').text.split('\n')[0])
        prd_url.append(li.find_element(By.CLASS_NAME,'baby-product-link').get_attribute('href'))
        # image 저장
        e = li.find_element(By.TAG_NAME,'img')
        img_url = e.get_attribute('src')
        img_res = requests.get(img_url)
        img_res.raise_for_status()

        food_name = './image_coupang/food{}.jpg'.format(idx+1)
        with open(food_name, "wb") as f:
            f.write(img_res.content)

        prd_foodname.append(food_name)
        if idx >= 5:
            break
        # for i in range(2,11):
        #     try:
        #         next_element.find_elements(By.TAG_NAME,'a')[i].cilck()
        #         driver.implicitly_wait(10)
        #         break
        #     except:
        #         pass
    except Exception:
        pass
    
```

```python
#import folium.map import Popup
path_Data_out = '../Data-out/'
import pandas as pd
import folium
```

```python
# 서울지도 만들기
seoul_map = folium.Map(location = [37.55, 126.98], zoom_start=12)

# 지도를 HTML 파일로 저장하기
seoul_map.save(path_Data_out + 'seoul.html')
```

```python
df = pd.read_excel('../Data-in/서울지역 대학교 위치.xlsx')
df.rename(
    columns = {'Unnamed: 0': '대학이름'},
    inplace=True
)
```

```python
for name, lat, long in zip(df.대학이름, df.위도, df.경도):
    university = folium.Marker([lat, long], popup = folium.Popup(name, max_width='300px')).add_to(seoul_map)
university.save(path_Data_out + 'seoul_university.html')

                                    
```

### IFrame popup

```python
for name, lat, long in zip(df.대학이름, df.위도, df.경도):
    iframe = folium.IFrame(name, width = 300, height =80)
    popup = folium.Popup(iframe, max_width=300)
    folium.Marker([lat, long], popup = popup).add_to(seoul_map)

seoul_map.save(path_Data_out + 'seoul_colleges.html')
seoul_map
```

### 원형 마커

```python
# 서울지도 만들기
seoul_map = folium.Map(location = [37.55, 126.98], zoom_start=12)

# 지도를 HTML 파일로 저장하기
seoul_map.save('../Data-in/seoul.html')
df = pd.read_excel('../Data-in/서울지역 대학교 위치.xlsx')
df.rename(
    columns = {'Unnamed: 0': '대학이름'},
    inplace=True
)
```

```python
for name, lat, long in zip(df.대학이름, df.위도, df.경도):
    iframe = folium.IFrame(name, width = 120, height =80)
    popup = folium.Popup(iframe, max_width=120)
    folium.CircleMarker([lat, long],
                        radius = 10,   #원의반지름
                        color = 'brown', # 원의 둘레 새상
                        fill = True,
                        fill_color = 'coral', # 원을 채우는 색
                        fill_opacity = 0.7,  # 투명도
                        popup=popup
    ).add_to(seoul_map)

seoul_map.save(path_Data_out + 'seoul_colleages2.html')
seoul_map
```
